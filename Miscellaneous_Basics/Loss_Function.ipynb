```python
import numpy as np
```

```python
y_predicted = np.array([1,1,0,0,1])
y_true = np.array([0.3, 0.7, 1,0,0.5])
```

```python
def mae(y_true, y_predicted):
    total_error = 0
    for yt, yp in zip(y_true, y_predicted):
        total_error += (abs(yt-yp))

    print(f"Total error is {total_error}")
    mae = total_error/len(y_true)
    print(f"MAE is {mae}")
    return mae
```

```python
mae(y_true, y_predicted)
```

```python
epsilon = 1e-15
epsilon
```

```python
y_predicted_new = [max(i,epsilon) for i in y_predicted]#replacing zero with a value very close to zero to remove log 0 issue
y_predicted_new
```

```python
y_predicted_new = [min(i,1-epsilon) for i in y_predicted_new]
y_predicted_new
```

```python
def log_loss(y_true, y_predicted):
    error = 0
    for yt, yp in zip(y_true, y_predicted):
        val = (yt*np.log(yp)) + ((1-yt) * np.log(1-yp))
        error += val
    cost = (-1/len(y_true)) * error
    print(f"Total error is {cost}")
    return cost
```

```python
log_loss(y_true, y_predicted_new)
```
