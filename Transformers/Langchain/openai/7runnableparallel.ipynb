```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnableSequence, RunnableParallel
from dotenv import load_dotenv
load_dotenv()
```

```python
llm = ChatOpenAI(model ='gpt-3.5-turbo')
```

```python
template1 = PromptTemplate(template = "Summarize the following text \n{text}",
                            input_variables = ['text'])
template2 = PromptTemplate(template = "Convert the following to German language \n{text}",
                            input_variables = ['text'])
```

```python
parser = StrOutputParser()
```

```python
parallel_chain = RunnableParallel(
    {
        "Summary" : RunnableSequence(template1, llm, parser),
        "German": template2 | llm | parser
    }
)
```

```python
text = "Artificial intelligence (AI) refers to computer systems designed to perform tasks that typically require human intelligence, such as learning, reasoning, and problem-solving. It powers everyday tools like recommendation engines, chatbots, and image recognition. AI is reshaping industries, improving efficiency, and sparking debates about ethics, transparency, and responsible use."
```

```python
result = parallel_chain.invoke({'text': text})
```
