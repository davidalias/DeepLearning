```python
from langchain_openai import ChatOpenAI
from dotenv import load_dotenv
load_dotenv()
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnableLambda, RunnablePassthrough, RunnableParallel
```

```python
llm = ChatOpenAI(model="gpt-3.5-turbo")
parser = StrOutputParser()
```

```python
template = PromptTemplate(template="Give me an article not more than 150 words about {topic}",
                          input_variables=['topic'])
```

```python
sequential_chain = template | llm | parser
```

```python
parallel_chain = RunnableParallel(
                  {
                    'Article':RunnablePassthrough(), 
                     'count': RunnableLambda(lambda article: len(article.split()))
                  }
                )
```

```python
final_chain = sequential_chain | parallel_chain
```

```python
result = final_chain.invoke({"topic":"Future jobs in AI"})
```
