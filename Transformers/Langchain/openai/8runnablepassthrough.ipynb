```python
from langchain_openai import ChatOpenAI
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import PromptTemplate
from langchain_core.runnables import RunnableParallel, RunnablePassthrough
from dotenv import load_dotenv
load_dotenv()
```

```python
llm = ChatOpenAI(model = 'gpt-3.5-turbo')
```

```python
template1 = PromptTemplate(template = "Generate 3 disadvantages of the topic \n {topic}",
                            input_variables = ['topic'])

template2 = PromptTemplate(template = "Generate a short Youtube script for \n {llm_output}",
                            input_variables = ['llm_output'])
```

```python
parser = StrOutputParser()
```

```python
sequential_chain = template1 | llm | parser
parallel_chain = RunnableParallel(
    {
        'Disadvantages': RunnablePassthrough(),
        'Script': template2 | llm | parser
    }
)
```

```python
final_chain = sequential_chain | parallel_chain
result = final_chain.invoke({'topic': "Excessive use of plastic materials"})
```
