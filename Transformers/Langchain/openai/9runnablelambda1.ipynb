```python
from langchain_core.runnables import RunnableLambda
```

```python
def word_count(article:str):
   """
   a list of words is generated, then length of the string is returned

   """
   return len(article.split())
```

```python
converted_runnable = RunnableLambda(word_count)
```

```python
count = converted_runnable.invoke("My name is Robert Downey Jr")
print(count)
```

#with lambda function -> single expression and returns something can be converted into a lambda function

```python
converted_runnable = RunnableLambda(lambda article: len(article.split()))
```

```python
count = converted_runnable.invoke("My name is Robert Downey Jr")
print(count)
```
