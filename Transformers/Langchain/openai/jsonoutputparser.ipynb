```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import JsonOutputParser
from dotenv import load_dotenv
load_dotenv()
```

```python
llm = ChatOpenAI(model = 'gpt-3.5-turbo')
```

```python
parser = JsonOutputParser()
template = PromptTemplate(template = "Give me details of the year, make and model of the car {car} \n {format_style}",
                            input_variables = ['car'],
                            partial_variables = {'format_style':parser.get_format_instructions()})#partial variables will not be filled during runtime, but before runtime
```

```python
#prompt = template.invoke({'car': 'Tata'})
#print(prompt)
```

```python
chain = template | llm | parser
result = chain.invoke({'car': 'Tata Nexon'})
```
