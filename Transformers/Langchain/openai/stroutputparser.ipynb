```python
from langchain_openai import ChatOpenAI
from dotenv import load_dotenv
load_dotenv()
from langchain_core.output_parsers import StrOutputParser #output as string
from langchain_core.prompts import PromptTemplate
```

```python
llm = ChatOpenAI(model = 'gpt-3.5-turbo')
template = PromptTemplate(template = "What is the advantage of {topic}",
                            input_variables = ['topic'])
parser = StrOutputParser()
```

```python
prompt = template.invoke({'topic':'regular exercise'})
print(prompt)
```

```python
result = llm.invoke(prompt)
```

```python
print(result)
```

```python
final_result = parser.invoke(result)
print(final_result)
```

USE OF CHAIN (INSTEAD OF USING INVOKE EVERYWHERE)

```python
chain = template | llm | parser
result = chain.invoke({'topic': 'Vegetarian Diet'})
```

```python
print(result)
```
