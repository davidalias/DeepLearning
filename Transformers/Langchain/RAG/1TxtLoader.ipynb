```python
from langchain_community.document_loaders import TextLoader
loader = TextLoader(file_path = 'text.txt', encoding ='utf-8')#utf-8 can encode unicode data/text and special data
```

```python
list_of_doc_objects = loader.load()#single object for text data. Each object has page content and metadata
```
