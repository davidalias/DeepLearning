```python
from langchain_chroma import Chroma
from langchain_openai import OpenAIEmbeddings
from dotenv import load_dotenv
load_dotenv()
from langchain_core.documents import Document
```

```python
embeddings = OpenAIEmbeddings(model="text-embedding-3-small")
```

```python
doc1 = Document(
    page_content="""Data Engineer with 5 years of experience in Python, SQL, and building ETL pipelines.
    Worked on cloud data warehouses like Snowflake and BigQuery. Skilled in Airflow scheduling.""",
    metadata = {'name': 'Alice', "email": 'abc@example.com', 'phone': '000000'}
)
doc2 = Document(
    page_content="""Software engineer with 6 years of experience in Java and SpringBoot.
    Built scalable web applications and microservices. Limited exposure to data engineering.""",
    metadata = {'name': "Darwin", 'email': 'xyz@example.com', 'phone':'123456'}
)
```

```python
list_of_docs = [doc1, doc2]
```

```python
vector_db = Chroma(collection_name="collection1", embedding_function=embeddings, persist_directory="dummyfolder")
```

```python
vector_db.add_documents(documents=list_of_docs)
```

```python
result = vector_db.get(include=['documents','metadatas'])
print(result)
```

##Similarity Search

```python
result = vector_db.similarity_search(
    query = "HR is looking for a data engineer with python and ETL experience", 
    k =1
    )
#k implies number of matching results
#this function first does embedding, then trigger chromadb to perform similarity search based on embedding and result is returned
```

```python
print(result)
```

```python
for individual in result:
    print("Name:", individual.metadata['name'], "Email:", individual.metadata['email'])
    print(individual.page_content)
```

```python
result_with_score = vector_db.similarity_search_with_score(
    query = "HR is looking for a data engineer with python and ETL experience", 
    k =1
)
```

```python
print(result_with_score)
```
