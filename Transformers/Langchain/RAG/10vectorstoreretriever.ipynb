```python
from langchain_chroma import Chroma
from langchain_openai import OpenAIEmbeddings
from langchain_core.documents import Document
from dotenv import load_dotenv
load_dotenv()
embeddings = OpenAIEmbeddings(model = 'text-embedding-3-small')
```

```python
doc1 = Document(
    page_content="""Data Engineer with 5 years of experience in Python, SQL, and building ETL pipelines.
    Worked on cloud data warehouses like Snowflake and BigQuery. Skilled in Airflow scheduling.""",
    metadata = {'name': 'Alice', "email": 'abc@example.com', 'phone': '000000'}
)
doc2 = Document(
    page_content="""Software engineer with 6 years of experience in Java and SpringBoot.
    Built scalable web applications and microservices. Limited exposure to data engineering.""",
    metadata = {'name': "Darwin", 'email': 'xyz@example.com', 'phone':'123456'}
)
```

```python
list_of_docs = [doc1, doc2]
```

```python
vector_db = Chroma.from_documents(
    documents=list_of_docs, 
    embedding=embeddings,
    collection_name="mycollection"
    )
```

```python
retriever = vector_db.as_retriever(search_kwargs = {'k':1})
```

```python
#vectorstore retriever is a runnable unlike similarity_search
result = retriever.invoke("Hiring data engineer with python and snowflake skills")
print(result)
```

```python
for i, doc in enumerate(result):
    print("Result:",i+1)
    print("-----------------------------------")
    print(doc.page_content)
    print(doc.metadata)
```
