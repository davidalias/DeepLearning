```python
from langchain_chroma import Chroma
from langchain_openai import OpenAIEmbeddings
from dotenv import load_dotenv
load_dotenv()
from langchain_core.documents import Document
```

```python
embeddings = OpenAIEmbeddings(model="text-embedding-3-small") #default model is "text-embedding-ada-002" with vector dims as 1536
#"text-embedding-3-small" has vector dim. of 1536 and is cheaper than more accurate "text-embedding-3-large" model
```

```python
list_of_docs = [
    Document(page_content="Hi, I am Davis", metadata={'country': 'India', 'city': 'Bangalore',  })
    ]
```

```python
vector_db = Chroma(collection_name="collection1", embedding_function=embeddings)
#persist_directory="FolderName" attribute will store embeddings in the disk, if not specified embeddings are stored in RAM
```

```python
vector_db.add_documents(documents= list_of_docs)
```

```python
result = vector_db.get()
print(result)#though embeddings is none here, internally embeddings are created
```

```python
result = vector_db.get(include=['documents','metadatas','embeddings'])
print(result)
```

```python
result = vector_db.get(include=['documents'])
print(result)
```
