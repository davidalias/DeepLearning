```python
from langchain_community.vectorstores import FAISS
from langchain_openai import OpenAIEmbeddings
from dotenv import load_dotenv
load_dotenv()
from langchain_core.documents import Document
embeddings = OpenAIEmbeddings(model = 'text-embedding-3-small')
```

```python
docs =[
    Document(page_content="AI assistants like Chatgpt help people to answer questions."),
    Document(page_content="Chatgpt is an AI assistant used for answering queries."),
    Document(page_content="AI chatbots are trained to provide conversational answers."),
    Document(page_content="Doctors use AI systems to analyze X-rays and detect diseases."),
    Document(page_content="Banks use AI for fraud detection and risk management."),
    Document(page_content="AI models help in predicting stock market trends.")
]
```

```python
vectore_store = FAISS.from_documents(documents=docs, embedding=embeddings)
```

```python
query = "How do AI assistants answer questions?"
```

##Maximal Marginal Relevance

```python
retriever_with_mmr = vectore_store.as_retriever(search_type="mmr", 
                           search_kwargs={"k": 3, "lambda_mult": 0.4}
                        )#lambda_mult = 1, only relevance is considered (works like normal similarity search)
#lambda_mult = 0, only diversity is considered (ignores query and try to pick documents that are as different as possible(not very useful))
```

```python
retriever_without_mmr = vectore_store.as_retriever( 
                           search_kwargs={"k": 3}
                        )
```

```python
result_with_mmr = retriever_with_mmr.invoke(query)
result_without_mmr = retriever_without_mmr.invoke(query)
```
