```python
from langchain_openai import OpenAIEmbeddings
from dotenv import load_dotenv
from langchain_chroma import Chroma
from langchain_core.documents import Document
```

```python
load_dotenv()
embeddings = OpenAIEmbeddings(model='text-embedding-3-small')
```

```python
vector_db= Chroma(collection_name='collection1', embedding_function=embeddings, persist_directory='dummyfolder')
```

```python
result = vector_db.get()
print(result)
```

```python
print(result['ids'])
```

```python
print(result['documents'])
```

```python
for id, doc in zip(result['ids'], result['documents']):
    print(id,":",doc)
```

##Update

```python
doc_update = Document(
    page_content="""Data Engineer with 5 years of experience in Python, SQL, and building ETL pipelines.
    Worked on cloud data warehouses like Snowflake and BigQuery. Skilled in Airflow scheduling and data science""",
    metadata = {'name':'Clint', 'email':'clint@example.com','phone':'1654253'}
)
```

```python
vector_db.update_document(document_id='d5134890-f841-44a0-a5f7-403aba12a214', document=doc_update)
```

```python
result = vector_db.get()
```

```python
for id, doc in zip(result['ids'], result['documents']):
    print(id,':',doc)
```
