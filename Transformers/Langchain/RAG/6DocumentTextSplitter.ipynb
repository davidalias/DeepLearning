```python
from langchain_text_splitters import RecursiveCharacterTextSplitter, Language
```

```python
#from_language will return splitter object which is based on specific language
splitter = RecursiveCharacterTextSplitter.from_language(language = Language.PYTHON, 
                                                        chunk_size = 300,
                                                        chunk_overlap = 0
                                                        )
```

```python
code = """
class Book:
    def __init__(self, title, author, year):
        self.title = title
        self.author = author
        self.year = year
        self.is_checked_out = False

class Library:
    def __init__(self, name):
        self.name = name
        self.books = []

    def find_book(self, title):
        for book in self.books:
            if book.title.lower() == title.lower():
                return book
        return None


# Example usage
if __name__ == "__main__":
    # Create library
    my_library = Library("City Library")

    # Create books
    book1 = Book("1984", "George Orwell", 1949)
    book2 = Book("To Kill a Mockingbird", "Harper Lee", 1960)


"""
```

```python
list_of_chunks = splitter.split_text(text = code)
```

```python
print(len(list_of_chunks))
```

```python
for chunk in list_of_chunks:
    print(chunk)
    print("-------------------")
```
