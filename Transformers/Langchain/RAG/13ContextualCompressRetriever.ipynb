```python
from langchain_community.vectorstores import FAISS
from langchain_openai import OpenAIEmbeddings, ChatOpenAI
from langchain_core.documents import Document
from dotenv import load_dotenv
load_dotenv()
from langchain_classic.retrievers.contextual_compression import ContextualCompressionRetriever
from langchain_classic.retrievers.document_compressors import LLMChainExtractor #extract refined text from shorlisted documents
```

```python
llm = ChatOpenAI(name='gpt-3.5-turbo')
embeddings = OpenAIEmbeddings(model = 'text-embedding-3-large')
```

```python
docs = [
    Document(
        page_content= """
        Langchain allows integration with multiple vector databases such as FAISS, Pinecone and Chroma.
        These databases store embeddings of documents for efficient sematic search.
        The Great wall of China is one of the seven wonders of the world.
        """
    ),
    Document(
        page_content= """ 
        In Langchain, retrievers use embeddings to fetch documents similar in meaning to the user's query.
        FAISS is an open-source library by Meta for fast similarity search on dense vectors.
        Python is a popular programming language used for AI development.
        """
    ),
    Document(
        page_content= """ 
        The contextual compression retriever in Langchain refines retrieved documents by removing irrelevant portions.
        It works well when the documents contain a mix of related and unrelated topics.
        The Himalayas are known for Mount Everest, the tallest mountain in the world.
        """
    ),
    Document(
        page_content= """ 
        Langchain provides easy adapters for connecting LLM applications with external data sources.
        One common use case in retrieval-augmented generation (RAG), where vector databases supply the knowledge base.
        Mumbai is the financial capital of India.
        """
    ),
    Document(
        page_content= """ 
        Langchain supports local storage of embeddings using FAISS for offline retrieval.
        Meanwhile, machine learning models benefit from large and clean datasets.
        Coffee beans are grown mainly in tropical regions.
        """
    ),
    Document(
        page_content= """ 
        The OpenAI API powers many Langchain integrations.
        The Eiffel Tower in Paris attracts millions of visitors every year.
        Vector databases like Pinecone are managed services, providing scalability and persistence.
        """
    )
]
```

```python
query = "How does Langchain use vector databases in its architecture?"
```

```python
vector_store = FAISS.from_documents(documents=docs, embedding= embeddings)
```

```python
baseRetriever = vector_store.as_retriever() #by default k =4
```

```python
compressor = LLMChainExtractor.from_llm(llm=llm)
```

```python
retriever = ContextualCompressionRetriever(base_retriever=baseRetriever, base_compressor=compressor)
```

```python
result = retriever.invoke(query)
```

```python
print(result)
```
