```python
import tensorflow as tf
```

```python
daily_sales_numbers = [21,22,-108,31,-1,32,34,31]
```

```python
tf_dataset = tf.data.Dataset.from_tensor_slices(daily_sales_numbers)
tf_dataset
```

```python
for sale in tf_dataset:
    print(sale)
```

```python
for sale in tf_dataset:
    print(sale.numpy())
```

```python
for sales in tf_dataset.take(3): #first 3 elements
    print(sales.numpy())
```

```python
tf_dataset = tf_dataset.filter(lambda x: x>0) #removing negative sale values
for sale in tf_dataset:
    print(sale.numpy())
```

```python
tf_dataset = tf_dataset.map(lambda x: x*91) #currency conversion from dollar to rupees
for sale in tf_dataset.as_numpy_iterator():
    print(sale)
```

```python
tf_dataset = tf_dataset.shuffle(2) #buffer size of 2 for picking random entries
for sale in tf_dataset.as_numpy_iterator():
    print(sale)
```

```python
for sales_batch in tf_dataset.batch(2):#helps to distribute these batches on different GPUs for training purpose
    print(sales_batch.numpy())
```

```python
#Everything in single line
tf_dataset = tf.data.Dataset.from_tensor_slices(daily_sales_numbers)
tf_dataset = tf_dataset.filter(lambda x: x>0).map(lambda y: y*91).shuffle(2).batch(2)

for sale_batch in tf_dataset.as_numpy_iterator():
    print(sale_batch)
```
